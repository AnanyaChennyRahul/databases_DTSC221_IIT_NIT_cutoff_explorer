```python
import tkinter as tk
from tkinter import ttk, messagebox
import mysql.connector
import tkintermapview  # Assuming you have this library installed
from PIL import Image, ImageTk

class ProgramExplorerApp:
    def __init__(self, master):
        self.master = master
        self.master.title("IIT-NIT Program Explorer")
        self.master.geometry("800x500")
       
        # Label for the title
        self.title_label = tk.Label(self.master, text="IIT-NIT Program Explorer", font=("Helvetica", 20))
        self.title_label.pack(pady=20)
       
        # Paragraph explaining the purpose
        self.purpose_text = "A user-friendly tool to help IIT/NIT aspirants select or find the institute best-suited for them. Use the FAQ button for commonly asked questions!, Understand your options using the Find my options button and see where different universities lie on the map!"
        self.purpose_label = tk.Label(self.master, text=self.purpose_text, wraplength=600, justify="center")
        self.purpose_label.pack(pady=10)
        
        # Slide show variables
        self.image_list = []  # List to store image paths
        self.image_index = 0  # Current image index
        self.image_label = tk.Label(self.master)  # Label to display slideshow image
        self.image_label.pack(pady=20)  # Position the label

        for image_path in["pics/iitbombay.jpeg","pics/iitkharagput.jpeg","pics/iitmadras.jpeg","pics/iitrourkee.jpeg", "pics/nitsurathkal.jpeg","pics/nittrichy.jpeg","pics/nitwarangal.jpeg"]:
            self.image_list.append(ImageTk.PhotoImage(Image.open(image_path)))
            
        # Start the slideshow loop
        self.show_slideshow()
              
        # Buttons for FAQ and MAP
        self.faq_button = ttk.Button(self.master, text="FAQ", command=self.open_faq_window)
        self.faq_button.pack(side=tk.LEFT, padx=20)
       
        self.map_button = ttk.Button(self.master, text="MAP", command=self.open_map_window)
        self.map_button.pack(side=tk.RIGHT, padx=20)

        self.find_rank_button = ttk.Button(self.master, text="Find My Options!", command=self.open_filter_window)
        self.find_rank_button.pack(pady=20)
        
        self.institute_details_button = ttk.Button(self.master, text="Institute Details", command=self.open_institute_details_window)
        self.institute_details_button.pack(padx=20)

        # Connect to the database
        self.connection = mysql.connector.connect(
            host="localhost",
            user="root",
            password="znmd",  # Change this to your password
            database="iit_nitcutoff"
        )
        self.cursor = self.connection.cursor()  # Create a cursor object
    
    def open_institute_details_window(self):
        # Functionality for opening Institute Details window
        institute_details_window = tk.Toplevel(self.master)
        institute_details_window.title("Institute Details")
        institute_details_window.geometry("400x300")

        # Create an instance of InstituteDetailsApp
        institute_app = InstituteDetailsApp(institute_details_window)

    def show_slideshow(self):
        self.image_label.config(image=self.image_list[self.image_index])
        self.image_index = (self.image_index + 1) % len(self.image_list)  # Wrap around
        self.master.after(3000, self.show_slideshow)
    def open_faq_window(self):
        # Functionality for opening FAQ window
        faq_window = tk.Toplevel(self.master)
        faq_window.title("FAQ")
        faq_window.geometry("800x600")
        
        faqs = [
            "What are the most popular programs offered ",
            "What is the average opening rank (for any program) required to get into an IIT in 2021?",
            "What is the cutoff for different quotas?",
            "What is the cutoff for different degrees?",
            "What is the cutoff for different categories?",
            "Which institutes are the most competitive?",
            "Which universities have the best NAAC grade vs NIRF ranking?",
            "What is the correlation between NAAC grade, year of establishment and campus size?",
            "What are the oldest IITs and NITs?",
            "Do bigger campuses have more events?",
            "Which universities have the most events?",
            "Which universities have no events?",
            "Which month has the most number of events?",
            "What time do events usually start at IITs and NITs?",
            "Which universities have the best work-event balance?",
            "What institutes have the highest enrollment?",
            "Does enrollment increase as campus size increases?",
            "Do bigger campuses have more events?",
            "Which institutes have the biggest campuses?",
            "Which universities have the most events per location?",
            "Over the COVID-19 lockdown period, which universities with 4+ year programs had events scheduled?"
        ]
        
        self.selected_faq = tk.StringVar()

        for idx, faq in enumerate(faqs):
            tk.Radiobutton(faq_window, text=faq, variable=self.selected_faq, value=str(idx)).pack(anchor='w')

        submit_button = tk.Button(faq_window, text="Submit", command=self.submit_faq)
        submit_button.pack(pady=10)
        
    def submit_faq(self):
        selected_index = self.selected_faq.get()
        if selected_index:
            selected_index = int(selected_index)
            queries = [
                # Query 1: What are the most popular programs offered 
                "SELECT program_name, COUNT(*) AS count FROM student_data GROUP BY program_name ORDER BY count DESC;",
                # Query 2: What is the average opening rank (for any program) required to get into an IIT in 2021?
                "SELECT AVG(opening_rank) AS avg_opening_rank, AVG(closing_rank) AS avg_closing_rank FROM student_data WHERE institute_type = 'IIT' AND Year = (SELECT MAX(Year) FROM student_data);",
                # Query 3: What is the cutoff for different quotas?
                "SELECT quota, AVG(opening_rank) AS avg_opening_rank, AVG(closing_rank) AS avg_closing_rank FROM student_data GROUP BY quota ORDER BY quota;",
                # Query 4: What is the cutoff for different degrees?
                "SELECT degree_short, AVG(opening_rank) AS avg_opening_rank, AVG(closing_rank) AS avg_closing_rank FROM student_data GROUP BY degree_short ORDER BY degree_short;",
                # Query 5: What is the cutoff for different categories?
                "SELECT category, AVG(opening_rank) AS avg_opening_rank, AVG(closing_rank) AS avg_closing_rank FROM student_data GROUP BY category ORDER BY category;",
                # Query 6: Which institutes are the most competitive? 
                "SELECT institute_name, institute_type, AVG(opening_rank) AS avg_opening_rank, AVG(closing_rank) AS avg_closing_rank FROM student_data GROUP BY institute_name, institute_type ORDER BY avg_opening_rank, avg_closing_rank ASC LIMIT 10;",
                # Query 7: Which universities have the best NAAC grade vs NIRF ranking?
                "SELECT Institute_Name, NAAC_Grade, NIRF_ranking FROM institute_details ORDER BY institute_details.NIRF_ranking ASC;",
                # Query 8: What is the correlation between NAAC grade, year of establishment and campus size?
                "SELECT institute_details.NAAC_Grade, institute_details.Established_Year, institute_details.Campus_size FROM institute_details;",
                # Query 9: What are the oldest IITs and NITs?
                "SELECT institute_name, institute_type, Established_Year FROM institute_details WHERE institute_type IN ('IIT', 'NIT') ORDER BY Established_Year ASC;",
                # Query 10: Do bigger campuses have more events?
                "SELECT institute_details.Institute_Name, institute_details.Campus_size, COUNT(*) AS event_count FROM institute_details INNER JOIN events ON institute_details.Uni_id = events.Uni_id GROUP BY institute_details.Institute_Name, institute_details.Campus_size ORDER BY institute_details.Campus_size ASC;",
                # Query 11: Which universities have the most events?
                "SELECT institute_details.Institute_Name, COUNT(*) AS event_count FROM institute_details INNER JOIN events ON institute_details.Uni_id = events.Uni_id GROUP BY institute_details.Institute_Name ORDER BY event_count DESC LIMIT 10;",
                # Query 12: Which universities have no events?
                "SELECT institute_details.Institute_Name FROM institute_details LEFT JOIN events ON institute_details.Uni_id = events.Uni_id WHERE events.Uni_id IS NULL;",
                # Query 13: Which month has the most number of events?
                "SELECT MONTH(Date) AS event_month, COUNT(*) AS event_count FROM events GROUP BY MONTH(Date) ORDER BY event_count DESC;",
                # Query 14: What time do events usually start at IITs and NITs?
                "SELECT TIME(TIME) AS START_TIME, COUNT(*) AS EVENT_COUNT FROM EVENTS WHERE UNI_ID IN( SELECT UNI_ID FROM INSTITUTE_DETAILS WHERE INSTITUTE_TYPE IN ('IIT', 'NIT')) GROUP BY TIME(TIME) ORDER BY EVENT_COUNT DESC; " ,
                # Query 15: Which universities have the best work-event balance?
                "SELECT institute_details.institute_name, institute_details.nirf_ranking, count(*) as event_count, (institute_details.NIRF_ranking/(1+ COUNT(*))) AS WEBS FROM institute_details INNER JOIN events ON institute_details.Uni_id=events.Uni_id GROUP BY institute_details.Institute_Name, institute_details.NIRF_ranking ORDER BY WEBS DESC LIMIT 10;",  
                # Query 16: What institutes have the highest enrollment?
                "SELECT Institute_Name, Total_enrollment FROM institute_details ORDER BY total_enrollment desc limit 10;",
                # Query 17: Does enrollment increase as campus size increases?
                "SELECT Institute_Name, Total_enrollment, campus_size FROM institute_details ORDER BY campus_size desc limit 10;",
                # Query 18: Do bigger campuses have more events?
                "SELECT institute_details.Institute_Name, institute_details.Campus_Size, COUNT(*) AS Number_Of_Events FROM institute_details INNER JOIN events ON institute_details.Uni_id = events.Uni_id GROUP BY institute_details.Institute_Name, institute_details.Campus_Size;",
                # Query 19: Which institutes have the biggest campuses?
                "SELECT Institute_Name, Campus_Size FROM institute_details ORDER BY Campus_Size DESC LIMIT 10;",
                # Query 20: Which universities have the most events per location?
                "SELECT Location, COUNT(*) AS Event_Count FROM institute_details INNER JOIN events ON institute_details.Uni_id = events.Uni_id GROUP BY Location ORDER BY Event_Count DESC LIMIT 10;",
                # Query 21: Over the COVID-19 lockdown period, which universities with 4+ year programs had events scheduled?
                "SELECT DISTINCT institute_details.Institute_Name, institute_details.Location, student_data.program_name FROM institute_details INNER JOIN events ON institute_details.Uni_id = events.Uni_id INNER JOIN student_data ON institute_details.Uni_id = student_data.Uni_id WHERE events.Date BETWEEN '2020-01-01' AND '2020-12-31' AND student_data.program_duration > 3;"
            ]

            query = queries[selected_index]
            result = self.execute_query(query)
            column_names = [desc[0] for desc in self.cursor.description]
            self.display_results(result, column_names)

    def execute_query(self, query):
                    self.cursor.execute(query)
                    result = self.cursor.fetchall()
                    return result

    def display_results(self, result, column_names):
                    answer_window = tk.Toplevel(self.master)
                    answer_window.title("FAQ Answer")
                    answer_window.geometry("600x400")

                    tree = ttk.Treeview(answer_window, columns=column_names, show='headings')  # Specify columns explicitly

                    # Configure columns (rest of the code remains the same)
                    for col in column_names:
                        tree.heading(col, text=col)
                        tree.column(col, anchor='center')

                    # Insert data
                    for row in result:
                        tree.insert("", tk.END, values=row)

                    tree.pack(expand=True, fill="both")
               


    def open_map_window(self):
        # Functionality for opening MAP window
        map_window = tk.Toplevel(self.master)
        map_window.title("IIT/NIT Map")
        map_window.geometry("900x700")

        my_label = ttk.LabelFrame(map_window)
        my_label.pack(pady=20)

        map_widget = tkintermapview.TkinterMapView(my_label, width=800, height=600, corner_radius=0)

        # Create a function to open a blank window when a marker is clicked
        def open_blank_window(name):
            blank_window = tk.Toplevel(map_window)
            blank_window.title(f"{name} Information")
            blank_window.geometry("400x300")
            label = tk.Label(blank_window, text=f"Information for {name}")
            label.pack(pady=20)
            print(f"Opened window for: {name}")  # Debug statement to check functionality

        # Add markers with the command to open blank windows
        locations = [
            (12.9915, 80.2337, "IIT Madras"),
            (28.5457, 77.1928, "IIT Delhi"),
            (26.4710, 73.1134, "IIT Jodhpur"),
            (22.3149, 87.3105, "IIT Kharagpur"),
            (19.1334, 72.9133, "IIT Bombay")
        ]
        for lat, lon, name in locations:
            map_widget.set_position(lat, lon, marker=True, text=name, command=lambda name=name: open_blank_window(name))  # Proper capture in lambda

        map_widget.set_zoom(5)
        map_widget.pack()
   
    def open_filter_window(self):
        filter_window = tk.Toplevel(self.master)
        filter_window.title("Find My Options!")
        filter_window.geometry("400x400")

        self.round_label = ttk.Label(filter_window, text="Choose Round:")
        self.round_combobox = ttk.Combobox(filter_window)
        self.round_label.pack(pady=5)
        self.round_combobox.pack(pady=5)

        self.institute_type_label = ttk.Label(filter_window, text="Choose Institute Type:")
        self.institute_type_listbox = ttk.Combobox(filter_window)
        self.institute_type_label.pack(pady=5)
        self.institute_type_listbox.pack(pady=5)

        self.institute_names_label = ttk.Label(filter_window, text="Choose Institute Name:")
        self.institute_names_listbox = ttk.Combobox(filter_window)
        self.institute_names_label.pack(pady=5)
        self.institute_names_listbox.pack(pady=5)

        self.program_names_label = ttk.Label(filter_window, text="Choose Program Name:")
        self.program_names_listbox = ttk.Combobox(filter_window)
        self.program_names_label.pack(pady=5)
        self.program_names_listbox.pack(pady=5)

        self.categories_label = ttk.Label(filter_window, text="Choose Category:")
        self.categories_listbox = ttk.Combobox(filter_window)
        self.categories_label.pack(pady=5)
        self.categories_listbox.pack(pady=5)

        self.round_combobox['values'] = self.get_rounds()
        self.round_combobox.bind("<<ComboboxSelected>>", self.update_institute_types)

        self.institute_type_listbox.bind("<<ComboboxSelected>>", self.update_institute_names)
        self.institute_names_listbox.bind("<<ComboboxSelected>>", self.update_program_names)
        self.program_names_listbox.bind("<<ComboboxSelected>>", self.update_categories)

        explore_button = ttk.Button(filter_window, text="Explore", command=self.explore_round)
        explore_button.pack(pady=5)

    def get_rounds(self):
        # Functionality for fetching rounds from the database
        self.cursor.execute("SELECT DISTINCT round_no FROM student_data order by round_no asc")
        rounds = [row[0] for row in self.cursor.fetchall()]
        return rounds

    def get_institute_types(self, selected_round):
        # Functionality for fetching institute types based on selected round from the database
        self.cursor.execute("SELECT DISTINCT institute_type FROM student_data WHERE round_no = %s", (selected_round,))
        institute_types = [row[0] for row in self.cursor.fetchall()]
        return institute_types

    def get_institute_names(self, selected_round, selected_institute_type):
        # Functionality for fetching institute names based on selected round and institute type from the database
        self.cursor.execute("SELECT DISTINCT institute_name FROM student_data WHERE round_no = %s AND institute_type = %s", (selected_round, selected_institute_type))
        institute_names = [row[0] for row in self.cursor.fetchall()]
        return institute_names

    def get_program_names(self, selected_round, selected_institute_type, selected_institute_name):
        # Functionality for fetching program names based on selected round, institute type, and institute name from the database
        self.cursor.execute("SELECT DISTINCT program_name FROM student_data WHERE round_no = %s AND institute_type = %s AND institute_name = %s", (selected_round, selected_institute_type, selected_institute_name))
        program_names = [row[0] for row in self.cursor.fetchall()]
        return program_names

    def get_categories(self, selected_round, selected_institute_type, selected_institute_name, selected_program_name):
        # Functionality for fetching categories based on selected round, institute type, institute name, and program name from the database
        self.cursor.execute("SELECT DISTINCT category FROM student_data WHERE round_no = %s AND institute_type = %s AND institute_name = %s AND program_name = %s", (selected_round, selected_institute_type, selected_institute_name, selected_program_name))
        categories = [row[0] for row in self.cursor.fetchall()]
        return categories

    def update_institute_types(self, event):
        selected_round = self.round_combobox.get()
        self.institute_type_listbox['values'] = self.get_institute_types(selected_round)
        self.institute_names_listbox.set("")  # Reset institute names
        self.program_names_listbox.set("")  # Reset program names
        self.categories_listbox.set("")  # Reset categories

    def update_institute_names(self, event):
        selected_round = self.round_combobox.get()
        selected_institute_type = self.institute_type_listbox.get()
        self.institute_names_listbox['values'] = self.get_institute_names(selected_round, selected_institute_type)
        self.program_names_listbox.set("")  # Reset program names
        self.categories_listbox.set("")  # Reset categories

    def update_program_names(self, event):
        selected_round = self.round_combobox.get()
        selected_institute_type = self.institute_type_listbox.get()
        selected_institute_name = self.institute_names_listbox.get()
        self.program_names_listbox['values'] = self.get_program_names(selected_round, selected_institute_type, selected_institute_name)
        self.categories_listbox.set("")  # Reset categories

    def update_categories(self, event):
        selected_round = self.round_combobox.get()
        selected_institute_type = self.institute_type_listbox.get()
        selected_institute_name = self.institute_names_listbox.get()
        selected_program_name = self.program_names_listbox.get()
        self.categories_listbox['values'] = self.get_categories(selected_round, selected_institute_type, selected_institute_name, selected_program_name)

    def explore_round(self):
        selected_round = self.round_combobox.get()
        selected_institute_type = self.institute_type_listbox.get()
        selected_institute_name = self.institute_names_listbox.get()
        selected_program_name = self.program_names_listbox.get()
        selected_category = self.categories_listbox.get()

        if selected_round:
            round_detail_window = tk.Toplevel(self.master)
            round_detail_window.title(f"Round Detail: {selected_round}")

            # Display round details in a treeview
            result_column = ("institute_name", "program_name", "quota", "category", "pool", "opening_rank", "closing_rank")
            tree = ttk.Treeview(round_detail_window, columns=result_column, show='headings')
            tree.heading("institute_name", text="Institute")
            tree.heading("program_name", text="Program")
            tree.heading("quota", text="Quota")
            tree.heading("category", text="Category")
            tree.heading("pool", text="Pool")
            tree.heading("opening_rank", text="Opening Rank")
            tree.heading("closing_rank", text="Closing Rank")

            query = f"""
                SELECT institute_name, program_name, quota, category, pool,
                    opening_rank, closing_rank
                FROM student_data
                WHERE round_no = {selected_round}
            """
            filters = []
            if selected_institute_type:
                filters.append(f"institute_type = '{selected_institute_type}'")
            if selected_institute_name:
                filters.append(f"institute_name = '{selected_institute_name}'")
            if selected_program_name:
                filters.append(f"program_name = '{selected_program_name}'")
            if selected_category:
                filters.append(f"category = '{selected_category}'")
            if filters:
                query += " AND " + " AND ".join(filters)

            self.cursor.execute(query)
            rows = self.cursor.fetchall()
            for row in rows:
                tree.insert("", "end", values=row)

            tree.pack(expand=True, fill="both")
    


class InstituteDetailsApp:

        # Your initialization code...
    def __init__(self, master):
            self.master = master
            self.master.title("Institute Details")
            self.master.geometry("400x300")

            self.institute_type = tk.StringVar()
            self.selected_institute = tk.StringVar()

            # Label and Combobox for selecting institute type
            self.institute_type_label = ttk.Label(self.master, text="Choose Institute Type:")
            self.institute_type_label.pack(pady=5)
            self.institute_type_combobox = ttk.Combobox(self.master, textvariable=self.institute_type, state="readonly", values=["IIT", "NIT"])
            self.institute_type_combobox.bind("<<ComboboxSelected>>", self.update_institutes)
            self.institute_type_combobox.pack(pady=5)

            # Label and Combobox for selecting institute
            self.institute_label = ttk.Label(self.master, text="Choose Institute:")
            self.institute_label.pack(pady=5)
            self.institute_combobox = ttk.Combobox(self.master, textvariable=self.selected_institute, state="readonly")
            self.institute_combobox.pack(pady=5)

            # Button to submit selection
            self.submit_button = ttk.Button(self.master, text="Submit", command=self.show_institute_details)
            self.submit_button.pack(pady=10)

            # Connect to the database
            self.connection = mysql.connector.connect(
                host="localhost",
                user="root",
                password="znmd",  # Change this to your MySQL password
                database="iit_nitcutoff"   # Change this to your database name
            )
            self.cursor = self.connection.cursor()
            ########

    def update_institutes(self, event):
            # Update institute combobox based on selected institute type
            selected_type = self.institute_type.get()
            if selected_type:
                self.cursor.execute("SELECT Institute_name FROM Institute_details WHERE Institute_type = %s", (selected_type,))
                institutes = [row[0] for row in self.cursor.fetchall()]
                self.institute_combobox["values"] = institutes

    def show_institute_details(self):
            # Display details of selected institute
            selected_institute = self.selected_institute.get()
            if selected_institute:
                query = """
                    SELECT Institute_name, Total_enrollment, Campus_size, NAAC_Grade, NIRF_ranking, Established_year
                    FROM Institute_details
                    WHERE Institute_name = %s
                """
                self.cursor.execute(query, (selected_institute,))
                result = self.cursor.fetchone()
                if result:
                    details_window = tk.Toplevel(self.master)
                    details_window.title("Institute Details")
                    details_window.geometry("400x250")

                    institute_name_label = ttk.Label(details_window, text=f"Institute Name: {result[0]}")
                    institute_name_label.pack(pady=5)

                    total_enrollment_label = ttk.Label(details_window, text=f"Total Enrollment: {result[1]}")
                    total_enrollment_label.pack(pady=5)

                    campus_size_label = ttk.Label(details_window, text=f"Campus Size: {result[2]} acres")
                    campus_size_label.pack(pady=5)

                    naac_grade_label = ttk.Label(details_window, text=f"NAAC Grade: {result[3]}")
                    naac_grade_label.pack(pady=5)

                    nirf_ranking_label = ttk.Label(details_window, text=f"NIRF Ranking: {result[4]}")
                    nirf_ranking_label.pack(pady=5)

                    established_year_label = ttk.Label(details_window, text=f"Established Year: {result[5]}")
                    established_year_label.pack(pady=5)

                    # Fetch and display number of events for the selected institute
                    events_query = "SELECT COUNT(*) FROM Events WHERE Uni_ID = (SELECT Uni_id FROM Institute_details WHERE Institute_name = %s)"
                    self.cursor.execute(events_query, (selected_institute,))
                    num_events = self.cursor.fetchone()[0]
                    events_label = ttk.Label(details_window, text=f"Number of Events per Year: {num_events}")
                    events_label.pack(pady=5)
                else:
                    messagebox.showinfo("Error", "No details found for the selected institute.")
            else:
                messagebox.showinfo("Error", "Please select an institute.")


# Create the main application window
root = tk.Tk()
app = ProgramExplorerApp(root)
root.mainloop()
```
